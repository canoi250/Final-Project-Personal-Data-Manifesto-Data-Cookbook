```python
from geopy.distance import geodesic
import pandas as pd
import requests
import matplotlib.pyplot as plt
import numpy as np 
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import statsmodels.api as sm
import seaborn as sns
```

```python
mortality_rate = pd.read_csv("youth-mortality-rate.csv")
```

```python
mortality_rate.head(100)
```

```python
germany = mortality_rate[mortality_rate['Entity'] == 'Germany']
russia = mortality_rate[mortality_rate['Entity'] == 'Russia']
afghanistan = mortality_rate[mortality_rate['Entity'] == 'Afghanistan']
algeria = mortality_rate[mortality_rate['Entity'] == 'Algeria']
vietnam = mortality_rate[mortality_rate['Entity']=="Vietnam"]
plt.figure(figsize = (25,10))
plt.scatter(germany['Year'], germany['Youth mortality rate (UN IGME (2021))'], color="brown", label="Germany")
plt.plot(germany['Year'], germany['Youth mortality rate (UN IGME (2021))'], color="brown")
plt.scatter(russia['Year'], russia['Youth mortality rate (UN IGME (2021))'], color="pink", label="Russia")
plt.plot(russia['Year'], russia['Youth mortality rate (UN IGME (2021))'], color="pink")
plt.scatter(afghanistan['Year'], afghanistan['Youth mortality rate (UN IGME (2021))'], color="blue", label="Afghanistan")
plt.plot(afghanistan['Year'], afghanistan['Youth mortality rate (UN IGME (2021))'], color="blue")
plt.scatter(algeria['Year'], algeria['Youth mortality rate (UN IGME (2021))'], color="red", label="Algeria")
plt.plot(algeria['Year'], algeria['Youth mortality rate (UN IGME (2021))'], color="red")

plt.xlabel('Year', fontsize=18)
plt.ylabel('Youth mortality rate (UN IGME (2021))', fontsize=18)
plt.title('Youth Mortality rates according to the UN', fontsize=18)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=18)
plt.show()
```
