```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the csv file with female labor force participation in Albania data 
flfp_albania = pd.read_csv("FLFP Data Albania - Sheet1.csv")
```

```python
flfp_albania
```

```python
# Create a new column with percentages to answer the question: What is the ratio of women who receive tertiary education and then join the workforce from 1990 to 2021?
flfp_albania['Tertiary Education to FLFP Ratio']= flfp_albania["Educ"]/flfp_albania["FLFP"]*100
```

```python
flfp_albania
```

Explanation: We first label the csv file flfp_albania using the pd.read_csv function. The next code creates a new column called "Tertiary Education to FLFP Ratio" in a dataframe called flfp_albania. The values in this new column are calculated by dividing the "Educ" column in the flfp_albania DataFrame by the "FLFP" column, and multiplying the result by 100 to get a percentage. This new column represents the ratio of women who receive tertiary education and then join the workforce in Albania, from 1990 to 2021.
